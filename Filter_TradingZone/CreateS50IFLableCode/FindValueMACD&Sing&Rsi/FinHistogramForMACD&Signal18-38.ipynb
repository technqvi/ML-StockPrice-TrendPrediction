```python
import pandas as pd  
import numpy as np  
import math as math

import plotly.express as px
import plotly.graph_objects as go
```

```python
#fileNamePath="rsi_macd_1H_SET50_Signal18And38.csv"
fileNamePath="S50F_MACD-RSI_1H.csv"
#fileNamePath="dataset/rsi_macd_1H_SET50-SignalAndMA-RSI_10.csv"

df_set50=pd.read_csv("dataset/"+fileNamePath,index_col='Date/Time',parse_dates =['Date/Time'],dayfirst=True)


df_set50.info()
```

```python
df_set50.tail()
```

```python
startPeriod='2012'
endPeriod='2015'
marketDefine='Big Tend 2y'
#macd_df=df_set50.loc[startPeriod:endPeriod,['MACD1','Signal1','Hist1','MACD2','Signal2','Hist2']]

#macd_df=df_set50.loc[startPeriod:endPeriod,['MACD2','Signal2','Hist2','Signal_2_Test25','Hist_2_Test25']]

macd_df=df_set50.loc[startPeriod:,['MACD1','Signal1','Hist1','MACD2','Signal2','Hist2','macd-d','signal-d','hist-d','macd-d2','signal-d2','hist-d2']]
des_macd_df=macd_df.describe(percentiles= [.25,.2, .5,.8 ,.75])
print(marketDefine,' : ',macd_df.index[0],' - ',macd_df.index[macd_df.shape[0]-1])
des_macd_df
```

```python
colNametoPlot='macd-d'
fig=px.histogram(macd_df,x=colNametoPlot,marginal="box",width=700,height=350)
fig.show()
```

```python
#fig = px.box(macd_df,y="MACD1")
fig=go.Figure()
fig.add_trace(
go.Box(y=macd_df[colNametoPlot],name=colNametoPlot,marker_color='red',boxmean=True)
)

fig.show()
```
