```python
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

```python
fileName="S50_2016_Now150x600.csv"
filePath="datasets"
s50_df=pd.read_csv(f'{filePath}\\{fileName}',index_col="Date/Time",parse_dates =['Date/Time','StartUp','StartDown'],dayfirst=True)
print(s50_df.shape)
```

```python
print(s50_df.isnull().sum())
```

```python
s50_df.drop(columns='Ticker',inplace=True)
s50_df.dropna(axis=0,how='any',inplace=True)
print("After remove Null: ",s50_df.shape)
```

```python
s50_df[['Uptrend','SumUp', 'BarUp','Downtrend','SumDown','BarDown']] = s50_df[['Uptrend','SumUp', 'BarUp','Downtrend','SumDown','BarDown']].apply(pd.to_numeric,downcast='integer')
```

```python
def ListDescribeByPeriod(start,end,cols):

 df_CountAB=s50_df.loc[start:end,cols]
 return df_CountAB.describe(percentiles=[.3,.50,.7])
```

```python
dfDes2019=ListDescribeByPeriod('01-2019','12-2019',['SumUp','SumDown'])
dfDes2019
```

```python
dfDes2016_R1=ListDescribeByPeriod('01-2016','08-2016',['SumUp','SumDown'])
dfDes2016_Sideway=ListDescribeByPeriod('09-2016','12-2016',['SumUp','SumDown'])

dfDes2017_R1=ListDescribeByPeriod('01-2017','03-2017',['SumUp','SumDown'])
dfDes2017_Sideway=ListDescribeByPeriod('04-2017','24-08-2017',['SumUp','SumDown'])
dfDes2017_R2=ListDescribeByPeriod('25-08-2017','31-01-2018',['SumUp','SumDown'])
```
