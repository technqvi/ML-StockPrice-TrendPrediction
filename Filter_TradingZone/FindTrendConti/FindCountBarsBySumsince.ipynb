```python
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

# Demo find maxBar in rally trend

```python
df2019_trendByMA=pd.read_csv('FindTrend-DataSet\TrendByMA550_Avg100_2019_tf15m.csv',index_col='Date/Time',parse_dates =['Date/Time'],dayfirst=True)

df2019_trendByMA.dropna(inplace=True)
df2019_trendByMA.drop(columns='Ticker',inplace=True)
```

```python
display('df2019_trendByMA.head()','df2019_trendByMA.tail()')
```

```python
dfCount_EndBarIncrease=df2019_trendByMA[ (df2019_trendByMA['countPerMA'].shift(-1)== df2019_trendByMA['countPerMA'])  & (df2019_trendByMA['countPerMA'].shift(1)!= df2019_trendByMA['countPerMA'])]
dfCount_EndBarIncrease
```

```python
dfCount_EndBarIncrease.groupby(['trendLevel550'],)[['countPerMA']].agg(['median','mean','max','min'])
```
